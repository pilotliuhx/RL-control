```python
import numpy as np
import matplotlib.pyplot as plt
G1, G2, G3, G4 = 0.863, -0.3135, 0.0186,  0.9968
H1, H2 = 0.3135, 0.00321
C1, C2 = 0, 1

x1log = []
x2log = []
ylog  = []
tlog = []
ulog = []
x1, x2 = 0.0, 0.0
u = 1
for i in range(100):
    x1_next = G1*x1 + G2*x2 + H1*u
    x2_next = G3*x1 + G4*x2 + H2*u 
    y = C1*x1 + C2*x2
    
    x1log.append(x1)
    x2log.append(x2)
    ylog.append(y)
    tlog.append(i*0.02)
    ulog.append(u)
    x1 = x1_next
    x2 = x2_next
plt.figure()
plt.plot(tlog,ulog, label='input')
plt.plot(tlog,ylog, label='output')
plt.legend()
plt.figure()
plt.plot(tlog,x1log, label='rate')
plt.plot(tlog,x2log, label='angle')
plt.legend()
```
