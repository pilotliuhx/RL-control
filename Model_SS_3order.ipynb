```python
import numpy as np
import matplotlib.pyplot as plt
G1, G2, G3 = 0.7483, -1.3184, 0.0
G4, G5, G6 = 0.0174,  0.9862, 0.0
G7, G8, G9 = 0.2881,  1.3345, 1.0

C1, C2 = 15.7272, 77.9572

H1, H2, H3 = 0.0174, 0.0002, 0.0030

x1log = []
x2log = []
x3log = []
rollratelog = []


x1, x2, x3 = 0.0, 0.0, 0.0
u = 1
for i in range(50):
    if i > 10 & i < 20:
        u = 1
    else:
        u = 0
    x1_next = G1*x1 + G2*x2 + G3*x3 + H1*u
    x2_next = G4*x1 + G5*x2 + G6*x3 + H2*u
    x3_next = G7*x1 + G8*x2 + G9*x3 + H3*u
    
    x1log.append(x1)
    x2log.append(x2)
    x3log.append(x3)
    
    x1 = x1_next
    x2 = x2_next
    x3 = x3_next
    
    rollrate = C1*x1 + C2*x2
    rollratelog.append(rollrate)
```

```python
plt.plot(x1log, label='x1')
plt.plot(x2log, label='x2')
plt.plot(x3log, label='x3')
plt.legend()
```
